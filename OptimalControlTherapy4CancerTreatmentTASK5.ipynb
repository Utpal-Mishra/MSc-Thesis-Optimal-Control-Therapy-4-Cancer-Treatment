### **Task 5: 2-Dosage Per Day Optimization Contour Plot**

```python
def pkmodel(x, t):
   CL   = 17.3446 
   ka   = 0.61 
   v    = 319
   chig = x[0]
   chib = x[1]

   dydt = [-ka*chig, ka*chig - CL*(chib/v)]
   return dydt

def get_penalty_fn(t, chib, v):
    C_targ = 0.57 
    temp   = []
    dt     = t[1] - t[0];
         
    for i in range(len(t)):
        z_val = chib[i]/v
        penalty_val = abs(z_val - C_targ)

        temp.append(penalty_val)
        
    penalty = []
      
    for i in range(len(t)):
        penalty.append(sum(temp[1:i-1])*dt)
          
    return penalty


# fig = make_subplots(rows=1, cols=3, specs=[[{}, {}, {}]], subplot_titles=("Time vs chib","", "Time vs Penalty"))
for d1 in range(100, 900, 100):
    D                 = d1
    TD                = 12
    n_doses           = 10
    ka                = 0.61
    f                 = 1
    initial_condition = [D*f, 0]
    CL                = 17.3446
    v                 = 319
    n_gridpoints      = 20

    t_vec    = [] #np.zeros((n_gridpoints, n_doses))
    chig_vec = [] #np.zeros((n_gridpoints, n_doses))
    chib_vec = [] #np.zeros((n_gridpoints, n_doses))


    for i in range(0, 10, 2):
      t_eval = np.linspace(i*TD, (i+1)*TD, 160)
      Y      = (odeint(pkmodel, initial_condition, t_eval))

      chig = Y[:, 0]
      chib = Y[:, 1]

      t_init  = t_eval[0]
      t_final = t_eval[-1]

      dt      = (t_final - t_init)/(n_gridpoints - 1)
      ti      = np.linspace(t_init, t_final, 160)
      
      chig_temp = scipy.interpolate.splrep(t_eval, chig, ti)
      chib_temp = scipy.interpolate.splrep(t_eval, chib, ti)

      t_vec.append(ti)
      chig_vec.append(chig_temp[1])
      chib_vec.append(chib[0:160:4])

      initial_condition = [chig[-1] + D*f, chib[-1]]
    
    for i in range(len(t_vec)):
      t_vec[i] = t_vec[i][0:160:4]
    t = []
    for i in t_vec:
      for k in i:
          t.append(k)

    chig = chig_vec

    chib = []
    for i in chib_vec:
      for k in i:
          chib.append(k)

    penalty = get_penalty_fn(t, chib, v)
    penalty[0] = 0

    fig = make_subplots(rows=1, cols=3, specs=[[{}, {}, {}]], subplot_titles=("Time vs chib","", "Time vs Penalty"))
    fig.add_trace(go.Scatter(x=t, y=np.divide(chib, v), name = "chig: change in amount of Imatinib in the gastrointestinal tract with time 0-12: D="+str(d1)+"mg"), row=1, col=1)
    fig.add_trace(go.Scatter(x=t, y=penalty, name = "chib: change in the amount of Imatinib in the blood with time 0-12: D="+str(d1)+"mg"), row=1, col=3)
    fig.update_layout(height=450, width=1200, title_text="Optimal Control Therapy for Cancer Treatment") 
    fig.update_xaxes(title_text = "t [hrs]", showline=True, linewidth=2, rangeslider_visible=True, linecolor='black', mirror=True, row = 1, col = 1)
    fig.update_yaxes(title_text = "chib", showline=True, linewidth=2, linecolor='black', mirror=True, row = 1, col = 1)
    fig.update_xaxes(title_text = "t [hrs]", showline=True, linewidth=2, rangeslider_visible=True, linecolor='black', mirror=True, row = 1, col = 3)
    fig.update_yaxes(title_text = "penalty", showline=True, linewidth=2, linecolor='black', mirror=True, row = 1, col = 3)
# fig.show()

    for d2 in range(100, 900, 100):
      D                 = d2
      TD                = 12
      n_doses           = 10
      ka                = 0.61
      f                 = 1
      initial_condition = [D*f, 0]
      CL                = 17.3446
      v                 = 319
      n_gridpoints      = 20

      t_vec    = [] #np.zeros((n_gridpoints, n_doses))
      chig_vec = [] #np.zeros((n_gridpoints, n_doses))
      hib_vec = [] #np.zeros((n_gridpoints, n_doses))


      for i in range(1, 11, 2):
        t_eval = np.linspace(i*TD, (i+1)*TD, 160)
        Y      = (odeint(pkmodel, initial_condition, t_eval))

        chig = Y[:, 0]
        chib = Y[:, 1]

        t_init  = t_eval[0]
        t_final = t_eval[-1]

        dt      = (t_final - t_init)/(n_gridpoints - 1)
        ti      = np.linspace(t_init, t_final, 160)
        
        chig_temp = scipy.interpolate.splrep(t_eval, chig, ti)
        chib_temp = scipy.interpolate.splrep(t_eval, chib, ti)

        t_vec.append(ti)
        chig_vec.append(chig_temp[1])
        chib_vec.append(chib[0:160:4])

        initial_condition = [chig[-1] + D*f, chib[-1]]
      
      for i in range(len(t_vec)):
        t_vec[i] = t_vec[i][0:160:4]
      t = []
      for i in t_vec:
        for k in i:
            t.append(k)

      chig = chig_vec

      chib = []
      for i in chib_vec:
        for k in i:
            chib.append(k)

      penalty = get_penalty_fn(t, chib, v)
      penalty[0] = 0

      # fig = make_subplots(rows=1, cols=3, specs=[[{}, {}, {}]], subplot_titles=("Time vs chib","", "Time vs Penalty"))
      fig.add_trace(go.Scatter(x=t, y=np.divide(chib, v), name = "chig: change in amount of Imatinib in the gastrointestinal tract with time 12-24: D="+str(d2)+"mg"), row=1, col=1)
      fig.add_trace(go.Scatter(x=t, y=penalty, name = "chib: change in the amount of Imatinib in the blood with time 12-24: D="+str(d2)+"mg"), row=1, col=3)
      fig.update_layout(height=450, width=1200, title_text="Optimal Control Therapy for Cancer Treatment") 
      fig.update_xaxes(title_text = "t [hrs]", showline=True, linewidth=2, rangeslider_visible=True, linecolor='black', mirror=True, row = 1, col = 1)
      fig.update_yaxes(title_text = "chib", showline=True, linewidth=2, linecolor='black', mirror=True, row = 1, col = 1)
      fig.update_xaxes(title_text = "t [hrs]", showline=True, linewidth=2, rangeslider_visible=True, linecolor='black', mirror=True, row = 1, col = 3)
      fig.update_yaxes(title_text = "penalty", showline=True, linewidth=2, linecolor='black', mirror=True, row = 1, col = 3)
    fig.show()
```

### **Task 5: 2-Dosage Per Day Optimization (Lennon's Approach) Contour Plot**

```python
def fix_all_parameters():
    TD                = 24
    n_doses           = 10
    ka                = 0.61
    f                 = 1
    CL                = 17.3446
    v                 = 319
    n_gridpoints      = 20
    return TD, n_doses, ka, f, CL, v, n_gridpoints

def pkmodel(x, t):
   CL   = 17.3446 
   ka   = 0.61 
   v    = 319
   chig = x[0]
   chib = x[1]

   dydt = [-ka*chig, ka*chig - CL*(chib/v)]
   return dydt


def get_penalty_fn(t, chib, v):
    C_targ = 1
    G     = 0.57 
    temp   = []
    #print(t[1], t[0])
    dt     = t[1] - t[0]
    # learning_rate = 0.01
    penalty_prev = 1
    penalty_curr = 0
    # z_val = 0

    for i in range(len(t)):
        z_val = chib[i]/v
        penalty_curr = penalty_curr -  0.001*abs(z_val - C_targ)*penalty_prev
        penalty_prev = penalty_curr  
        temp.append(penalty_curr)
          
    penalty = []
    for i in range(len(t)):
        penalty.append(sum(temp[1:i-1])*dt)
          
    return penalty


D1 = np.arange(100, 900, 100)
penalty_vec = []
plt.figure()

for k in range(len(D1)):

    D_init                = (np.zeros(10) + D1[k])[0]
    # penalty_vec = []

    D2 = np.arange(100, 900, 100)
    for x in range(len(D2)):

      TD, n_doses, ka, f, CL, v, n_gridpoints = fix_all_parameters()

      t_vec    = []
      chig_vec = []
      chib_vec = []    

      D_final                 = (np.zeros(10) + D1[x])[0]
      initial_condition = [D_init*f, 0]
      
      for i in range(n_doses-1):
        t_eval = np.linspace(i*TD, (i+1)*TD, 80)
        Y      = (odeint(pkmodel, initial_condition, t_eval))

        chig = Y[:, 0]
        chib = Y[:, 1]
        t_init  = t_eval[0]
        t_final = t_eval[-1]

        dt      = (t_final - t_init)/(n_gridpoints - 1)
        ti      = np.linspace(t_init, t_final, 80)

        chig_temp = scipy.interpolate.splrep(t_eval, chig, ti)
        chib_temp = scipy.interpolate.splrep(t_eval, chib, ti)

        t_vec.append(ti)
        chig_vec.append(chig_temp[1])
        chib_vec.append(chib[0:80:4])
        initial_condition = [chig[-1] + D_final*f, chib[-1]]
            

      i = n_doses
      t_eval = np.linspace(i*TD, (i+1)*TD, 80)
      Y      = (odeint(pkmodel, initial_condition, t_eval))

      chig = Y[:, 0]
      chib = Y[:, 1]

      t_init  = t_eval[0]
      t_final = t_eval[-1]

      dt      = (t_final - t_init)/(n_gridpoints - 1)
      ti      = np.linspace(t_init, t_final, 80)
      
      chig_temp = scipy.interpolate.splrep(t_eval, chig, ti)
      chib_temp = scipy.interpolate.splrep(t_eval, chib, ti)

      t_vec.append(ti)
      chig_vec.append(chig_temp[1])
      chib_vec.append(chib[0:80:4])

      for i in range(len(t_vec)):
        t_vec[i] = t_vec[i][0:80:4]
      t = []
      for i in t_vec:
        for k in i:
            t.append(k)

      chig = chig_vec

      chib = []
      for i in chib_vec:
        for k in i:
            chib.append(k)

      penalty = get_penalty_fn(t, chib, v)
      penalty[0] = 0
      penalty_vec.append(sum(penalty))
      # print(D_final)
      # print(D_init)
      # print(penalty_vec, "\n")

      # fig = go.Figure(data = go.Scatter(x = D_eval, y = penalty_vec, name = "2-Dose Optimization Line"))
    #   fig.add_trace(go.Scatter(x=D1, y=penalty_vec, name = "Second Dosage: "+str(int(D1))+"mg"))
    #   fig.update_layout(height=450, width=800, title_text="Optimized 2-Dose Administration (Drug Dose "+str(int(D))+" in every 12 hrs)") 
    #   fig.update_xaxes(title_text = "Administered Drug Dose (mg)", showline=True, linewidth=2, rangeslider_visible=True, linecolor='black', mirror=True)
    #   fig.update_yaxes(title_text = "Penalty Function", showline=True, linewidth=2, linecolor='black', mirror=True)
    # fig.show()
```

```python
def optimal_dosage(penalty_new):
  for col in range(8):
    for row in range(8):
      if penalty_new[row][col] == np.min(penalty_new):
        return 100*(row+1), 100*(col+1)

penalty_new = np.array(penalty_vec).reshape(8, 8)
```

```python
X = np.arange(100, 900, 100)
Y = np.arange(100, 900, 100)
fig = go.Figure(data = go.Contour(x=X, y=Y, z = np.array(penalty_vec).reshape(8, 8), colorscale = "Electric"))#contours_coloring='heatmap', order = "F"))
fig.update_layout(height=500, width=800, title_text="Optimized Drug Administration for 2 Different Dosages in 48 hours") 
fig.update_xaxes(title_text = "Dose D2")
fig.update_yaxes(title_text = "Dose D1")
fig.show()
```
